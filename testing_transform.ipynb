```python
#OtoDomData clean
OtoDomData = OtoDomData.drop(['building_material', 'media_types', 'security_types', 'windows_type', 'construction_status', 'outdoor'], axis=1)
OtoDomData['furniture'] = None

OtoDomData['date'] = pd.to_datetime(OtoDomData['date'])
OtoDomData['create_date'] = pd.to_datetime(OtoDomData['create_date'].str.split('T').str[0])
OtoDomData['modify_date'] = pd.to_datetime(OtoDomData['modify_date'].str.split('T').str[0])

OtoDomData['lift'] = OtoDomData['lift'].map({'::y':1, '::n':0})
OtoDomData['car'] = OtoDomData['car'].str.replace('extras_types-85::garage', 1).fillna(0)
OtoDomData['rent'] = OtoDomData['rent'].str.replace(' zł', '').str.replace(' ','')
OtoDomData['floor'] = OtoDomData['floor'].str.replace('ground_floor', 'floor_0').str.split('_').str[-1]
OtoDomData['heating'] = OtoDomData['heating'].str.split('::').str[-1]
```

```python
#OlxData cleaning
OlxData = OlxData.rename(columns={'city_name':'city'})
OlxData[['building_year', 'car', 'heating', 'lift', 'rent']] = None

OlxData['date'] = pd.to_datetime(OlxData['date'])
OlxData['create_date'] = pd.to_datetime(OlxData['create_date'].str.split('T').str[0])
OlxData['modify_date'] = pd.to_datetime(OlxData['modify_date'].str.split('T').str[0])

OlxData['floor'] = OlxData['floor'].str.split('_').str[1]
OlxData['furniture'] = OlxData['furniture'].map({'yes':1, 'no':0})

OlxData['market_type'] = OlxData['market_type'].str.upper()
```

```python
word_to_num = {
    'one': 1,
    'two': 2,
    'three': 3,
    'four': 4,
    'five': 5,
    'six': 6,
    'seven': 7,
    'eight': 8,
    'nine': 9,
    'ten': 10
}

# Convert words to numbers
OlxData['rooms_num'] = OlxData['rooms_num'].map(word_to_num)
```
